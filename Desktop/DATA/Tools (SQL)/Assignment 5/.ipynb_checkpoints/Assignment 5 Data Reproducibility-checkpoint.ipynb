```python
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = sorted(glob.glob('data*.csv'))
dataset
```

```python
def visualize(filename) :
    plt.figure(figsize=(10,7))
    indexes = np.arange(0,40,5)
    labels = np.arange(0.40,5)
    features = np.arange(0, 40, 1)
    
    #Reading the .csv file 
    data = pd.read_csv(filepath_or_buffer = filename)
    
    #Finding the maximum
    maximum=data.apply(np.max)
    plt.plot(features, maximum, label='Max')
    
    #Finding the minimum
    minimum=data.apply(np.min)
    plt.plot(features, minimum, label='Min')
    
    #Finding the average
    average=data.apply(np.mean)
    plt.plot(features, average, label='Avg')

    plt.legend()
    #plt.title(filename)
    plt.xlabel('Features')
    plt.ylabel('Values')
    
    #Title of the graph w.r.t the dataset
    if filename == 'data-01.csv' :
        plt.title('Dataset #1')
        
    elif filename == 'data-02.csv' :
        plt.title('Dataset #2')
    
    elif filename == 'data-03.csv' :
        plt.title('Dataset #3')
        
    elif filename == 'data-04.csv' :
        plt.title('Dataset #4')
        
    plt.show()
```

```python
def identify_issues(filename) :
    
    data = pd.read_csv(filepath_or_buffer = filename)
    #Finding the minimum
    minimum=data.apply(np.min)
    #Assigning the summation of the minimum to a variable
    f = sum(minimum)
    #Maximum of feature#0 and feature#20
    m_feature0=data.iloc[:,0].max()
    m_feature20=data.iloc[:,20].max()
    #loop to check whether the max of feature0=0 and max of feature20=20
    if m_feature0 == 0 and m_feature20 == 20 :
        print("I doubt the integrity of the dataset")
    #loop to check whether the summation of minimum is equal to 0
    else :
        if f == 0 :
            print('Why is this happening?')
        else :
            print('Dataset looks good.')
```

```python
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    #calling the functions
    visualize(datasets)
    identify_issues(datasets)
```
